## Plot Choices and Schedules to See if they line up 

```python
def plot_behavioral_time_courses(sub,pain_or_reward='pain'):
    meta_data = pd.read_csv(analysisdir+'/Subject_Overview.csv')
    plt.figure(figsize=(12,3))
    
    if pain_or_reward=='pain':
        sub_trials = pd.DataFrame.from_csv(savedir+'behavioral_trial_table_'+sub+'_pain.csv',header=0,index_col=None)
        plt.title(sub+' '+str(meta_data['pain order'][meta_data['sub']==sub]))
        plt.plot(sub_trials['probabilities'],label='probability green is shocked')
        plt.plot(sub_trials['painchoice']-.05,linestyle='',marker='.',markersize=8,label='you chose green') # Green
        plt.plot(sub_trials['shock_del']-.15,linestyle='',marker='.',markersize=8,label='you got shocked') #reD
        plt.ylim(0,1)
        plt.legend(loc=3)
    elif pain_or_reward=='reward':
        sub_trials = pd.DataFrame.from_csv(savedir+'behavioral_trial_table_'+sub+'_rew.csv',header=0,index_col=None)
        plt.title(sub+' '+str(meta_data['reward order'][meta_data['sub']==sub]))
        plt.plot(sub_trials['probabilities'],label='probability green is rewarded')
        plt.plot(sub_trials['rewchoice']-.05,linestyle='',marker='.',markersize=8,label='you chose green') # Green
        plt.plot(sub_trials['rew_del']-.15,linestyle='',marker='.',markersize=8,label='you got rewarded') #reD
        plt.ylim(0,1)
        plt.legend(loc=3)
```

```python
plot_behavioral_time_courses('cb5','reward')
```

```python
def running_average(x,avgnum):
    smoothed = np.zeros((len(x),1))
    xcount=avgnum
    avgnum=float(avgnum)
    for xx in range(int(len(x)-avgnum)):
        xx = xx+avgnum
        smoothed[xx] = np.nansum(x[xcount-avgnum:xcount])/avgnum
        xcount=xcount+1

    return smoothed
```

```python
for sub in sublist_pain:
    sub_trials = pd.DataFrame.from_csv(savedir+'behavioral_trial_table_'+sub+'_pain.csv',header=0,index_col=None)
    plt.figure()
    plt.plot(running_average(sub_trials['paininformation'].as_matrix(),3))
    plt.plot(sub_trials['probabilities'].as_matrix())
```

```python
for sub in sublist_pain:
    sub_trials = pd.DataFrame.from_csv(savedir+'behavioral_trial_table_'+sub+'_rew.csv',header=0,index_col=None)
    plt.figure()
    plt.plot(running_average(sub_trials['rewinformation'].as_matrix(),3))
    plt.plot(sub_trials['probabilities'].as_matrix())
```
