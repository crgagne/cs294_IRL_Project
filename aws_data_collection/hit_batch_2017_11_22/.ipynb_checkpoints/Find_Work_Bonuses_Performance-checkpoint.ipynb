```python
ptable_ambi = ptable_ambi.loc[ptable_ambi['AID']!='1',]
ptable_ambi = ptable_ambi.loc[ptable_ambi['MID']!='chrisg',]
ptable_ambi = ptable_ambi.loc[ptable_ambi.date=='2016-11-21',]
```

```python
g= ptable_ambi['no_brainer_per_cor_gain'].as_matrix()
l = ptable_ambi['no_brainer_per_cor_loss'].as_matrix()
ptable_ambi['no_brainer_max'] = np.nanmax(np.vstack((g,l)),axis=0)
# bcause they only did one task, take the non-nan one
ptable_ambi_sorted = ptable_ambi.sort_values('no_brainer_max')
ptable_ambi_sorted = ptable_ambi_sorted.loc[~np.isnan(ptable_ambi_sorted['no_brainer_max']),]
```

```python
n= len(ptable_ambi_sorted)
print('number of mturkers')
print(n)
print('tiers')

t1=int(round(0.95*n))
t2=int(round(0.90*n))
t3=int(round(0.50*n))
print(t1)
print(t2)
print(t3)
```
