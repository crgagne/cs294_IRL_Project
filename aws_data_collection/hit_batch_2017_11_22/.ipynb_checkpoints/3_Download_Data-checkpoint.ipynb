```python
import datetime
## month to include 
month_greater_than_equal_to=3

## Connect to DB
db=_mysql.connect(host="christest.c5s5wgavsyu2.us-west-2.rds.amazonaws.com",user="christest",
                  passwd="gags1313",db="chrisdb")


## Get Header Names
db.query("""SELECT column_name FROM information_schema.columns WHERE table_name = 'web_experiments_ambi_trial'""")
r=db.store_result() # sends the whole results
header = r.fetch_row(0)
header = [hh[0] for hh in header]
print(header)


## Get Trials for Each Participant 
task='combined'
for MID,HID in zip(df_completed_bandit.MID,df_completed_bandit.HID):
    db.query("""SELECT * FROM web_experiments_ambi_trial WHERE (MID = '"""+str(MID)+"""') AND (loss_or_reward = '"""+task+"""') AND (HID = '"""+HID+"""')""")
    r=db.store_result() # sends the whole results
    rows = r.fetch_row(400)
    if len(rows)>0:
                        ## Write to CSV 
        with open(os.path.basename(os.getcwd())+task+'_'+MID+'.csv', 'w') as csvfile:
            writer = csv.writer(csvfile, delimiter=',',quotechar='"', quoting=csv.QUOTE_MINIMAL)
            writer.writerow(header)
            for row in rows:
                ## check for data
                if datetime.datetime.strptime(row[3], '%Y-%m-%d %H:%M:%S').month>=month_greater_than_equal_to:
                    writer.writerow([x for x in row])
```
