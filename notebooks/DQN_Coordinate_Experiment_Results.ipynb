```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import glob
%matplotlib inline
```

```python

def plot(folder,ugh=0):
    time_steps = np.loadtxt(glob.glob('../dqn/'+folder+'/gym/time_steps*')[0])
    mean_returns = np.loadtxt(glob.glob('../dqn/'+folder+'/gym/*mean*')[ugh])
    best_returns = np.loadtxt(glob.glob('../dqn/'+folder+'/gym/*best_mean*')[0])
    df=pd.DataFrame({'time_steps':time_steps,'mean_returns':mean_returns,'best_returns':best_returns})
    df1 = df.set_index('time_steps',drop=True)
    columns = [col for col in df1.columns if 'mean' in col]

    plt.figure(figsize=(8,3))
    sns.set_style('white')
    df1[columns].plot(linewidth=1,figsize=(8,3))
    plt.title('DQN ')
    #plt.axhline(y=20,linestyle='--',color='k',linewidth=1)
    sns.despine()
    plt.ylabel('mean returns')
    plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
    plt.xlabel('steps')
```
