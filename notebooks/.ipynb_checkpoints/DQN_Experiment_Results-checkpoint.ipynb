```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

# 11/21 - Running DQN on Presumably Smaller Network

- I'm pretty sure I set the network to be 1 convolutional layer of 16 filters of size 4. 
- And the other one, I had convolutional layers. 


```python
folder= 'cq_test'
```

```python
date = '2017-11-21-21:32'
time_steps = np.loadtxt('../dqn/'+folder+'/gym/time_steps'+date+'.txt')
mean_returns = np.loadtxt('../dqn/'+folder+'/gym/mean_episode_rewards'+date+'.txt')
best_returns = np.loadtxt('../dqn/'+folder+'/gym/best_mean_episode_rewards'+date+'.txt')
df=pd.DataFrame({'time_steps':time_steps,'mean_returns 1 layer':mean_returns,'best_returns 1 layer':best_returns})
df1 = df.set_index('time_steps',drop=True)
```

```python
# load in data
date = '2017-11-21-16:04'
time_steps = np.loadtxt('../dqn/'+folder+'/gym/time_steps'+date+'.txt')
mean_returns = np.loadtxt('../dqn/'+folder+'/gym/mean_episode_rewards'+date+'.txt')
best_returns = np.loadtxt('../dqn/'+folder+'/gym/best_mean_episode_rewards'+date+'.txt')
df=pd.DataFrame({'time_steps':time_steps,'mean_returns 2 layers':mean_returns,'best_returns 2 layers':best_returns})
df2 = df.set_index('time_steps',drop=True)
```

```python
date = '2017-11-21-14:23'
time_steps = np.loadtxt('../dqn/'+folder+'/gym/time_steps'+date+'.txt')
mean_returns = np.loadtxt('../dqn/'+folder+'/gym/mean_episode_rewards'+date+'.txt')
best_returns = np.loadtxt('../dqn/'+folder+'/gym/best_mean_episode_rewards'+date+'.txt')
df=pd.DataFrame({'time_steps':time_steps,'mean_returns 3 layers':mean_returns,'best_returns 3 layers':best_returns})
df3 = df.set_index('time_steps',drop=True)
```

```python
df_combined =df1.join(df2,how='outer')
df_combined =df_combined.join(df3,how='outer')
```

```python
columns = [col for col in df_combined.columns if 'mean' in col]
```

```python
plt.figure(figsize=(8,3))
sns.set_style('white')
df_combined[columns].plot(linewidth=1,figsize=(8,3))
plt.title('DQN on Pong using images \n (Different Exploration schedules) \n mean returns')
plt.axhline(y=20,linestyle='--',color='k',linewidth=1)
sns.despine()
plt.ylabel('mean returns')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.xlabel('steps')

#plt.savefig('figs/Q2_b.png',dpi=300)
```
